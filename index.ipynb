Covid19の分析

```kotlin
%use kotlin-statistics
```

```kotlin
%use lets-plot, krangl, klaxon
```

```kotlin
val df = DataFrame.readCSV("data/COVID-19.csv")
df.head()
```

```kotlin
val points = geom_point(
    data = mapOf(
        "x" to df["確定日"].asStrings().toList(),
        "y" to df["キー"].asStrings().toList()    
    ), alpha=1.0, sampling=sampling_pick(1000))
{
    x = "x" 
    y = "y"
}

ggplot() + points + ggsize(1000, 1000)
```

### 国内の発症者数の推移

```kotlin
import jetbrains.letsPlot.sampling.*

val eachCount = df["確定日"].asStrings().toList().groupingBy { it }.eachCount()
val listData = mapOf<String, List<String>>(
    "name" to eachCount.keys.map { it.toString() }.toList(),
    "count" to eachCount.values.map { it.toString() }.toList()
)

// グラフの描画
var p = lets_plot(listData)
p +=  geom_bar(stat = Stat.identity, sampling=sampling_pick(1000)) {
  x = "name"
  y = "count"
}
p + ggsize(1000, 500)
```
